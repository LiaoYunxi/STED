# 单细胞文件准备和预处理 #

```python
import os
import sys
import pandas as pd
import numpy as np
import STED
```

```python
data_dir = './demo/PBMC'
out_dir = './demo/PBMC/test_CorEx'
sc_count_file = os.path.join(data_dir,"10k_PBMC_Multiome_RNA_count.h5ad")
sc_anno_file = os.path.join(data_dir,"MainCelltype.txt")

seed = 1
ntopics = 10
```

gene_use参数配置用于基因筛选，gene_use 指定筛选方法（可选值为 'All'、'MARKER'、'HVG'，默认值为 'HVG'），其中使用 'MARKER' 方法时通过 ntopMarker（默认 200）指定筛选的顶级标记基因数量，使用 'HVG' 方法时通过 ntopHvg（默认 500）指定筛选的高可变基因数量。指定save_path会将gene_use的名单保存到该路径下。gene_use也接受list对象传入，直接指定用以后续分析的基因。

# 使用scTopic提取topic #

```python
# 使用CorEx模型推荐的数据预处理参数设置
scp.preprocessing(linear2log=True)

# 使用LDA模型推荐的数据预处理参数设置
# scp.preprocessing(linear2log=False)
```

anchored_genes支持嵌套list, 每个子list里面是一个topic的anchor gene

```python
# anchor_words = []# 无监督
anchor_words = [
    ["CD79A", "MS4A1"],
    ["GNLY", "NKG7"],
    ["CCL5", "GZMK"],
   ["IL7R", "RORA","TCF7", "LEF1"],
   ["FCGR3A", "MS4A7", "LYZ", "CD14"],
   ["FCER1A", "CST3"]
]
```

```python
scp.preForGuide(anchored_genes=anchor_words)
```

构建scTopic对象，用以模型训练

```python
sct = scTopic()
sct.setData(scp,out_dir)
```

设置hierarchy_topic来指导层次建模，如当ntopics为10，hierarchy_topic为[6,2]时，第一次CorEx构造10个主题，第二次在10个主题基础上再构造6个，第三次为在6个主题上再构建2个大主题。tc_plot设置为True时，图片保存在out_dir/model/文件夹下面。

```python
sct.trainCorEx(seed=seed,n_iter=200,n_ensemble=0,ntopics=ntopics,
               anchor_strength=2,hierarchy_topic = [6,2],tc_plot=True)
```

# 使用scTopic+epiDecon估计样本的细胞比例和表观信号 #

构建epiDecon对象，用以反卷积和信号预测。out_dir为结果保存路径，需要预处理后的单细胞数据gsPreProcessing对象和bulk数据gsPreProcessing对象。

```python
epi = epiDecon()
epi.SetData(out_dir,scp,epid)
```

model可以输入训练好的scTopic模型，如果没有输入，model=None，epiDecon会自动检索out_dir路径下文件。
设置benchmark=True,epiDecon保存反卷积中的过程文件。

如果model指定，epiDecon自动根据scTopic训练时选择出的更好的标准化方式进行反卷积；如果model为None，需要指定对于topic_celltype相关性矩阵的标准化方式，支持"Raw","Norm","NormBySD","Bayes","BayesNorm"(默认)。


Raw：直接使用topic_celltype原始权重进行计算。
Norm,对topic_celltype原始权重 列求和归一化 (Sum-Norm),将每个主题下的细胞类型权重归一化（和为1）。这将其视为概率分布：即“给定该主题，是某细胞类型的概率”。
NormBySD,使用 StandardScaler(with_mean=False)。通过标准差缩放权重，消除不同主题间数值量纲的差异，但不移除均值。
Bayes，使用贝叶斯方法生成的celltype_topic_mat_bayes.txt矩阵文件，不做额外处理。
BayesNorm,使用贝叶斯方法生成的celltype_topic_mat_bayes.txt矩阵文件，但对其进行列归一化（每个主题的权重和为1）。

```python
epi.Decon(ntopics_selection=ntopics,seed_selection=seed,
          model=sct,model_selection="CorEx",#"LDA","ALDA"(guided LDA)
          outname=None,benchmark=False)
epi.Bayes()
```

预测出的细胞比例保存在gsPreProcessing对象的celltype_frac_df属性中。
