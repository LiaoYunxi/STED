# Single-cell Data Preparation and Preprocessing #

```python
import os
import sys
import pandas as pd
import numpy as np
import time

from STED.Preprocessing import *
```

```python
data_dir = '/data/lyx/LDA/test_ipynb/demo'
out_dir = '/data/lyx/LDA/test_ipynb/demo/test_BERTopic'

sc_count_file = os.path.join(data_dir,"10k_PBMC_Multiome_RNA_counts.h5ad")
sc_anno_file = os.path.join(data_dir,"MainCelltype.txt")
gene_anno_file = os.path.join(data_dir,"OS_scRNA_gene_index.19264.tsv")
pretrained_model_file = os.path.join(data_dir,"models.ckpt")

gs_count_file = os.path.join(data_dir,"peaks_GAS.txt")

seed = 1
resolution = 0.5
fc = 0.0025
```

```python
if not os.path.exists(out_dir):
        os.makedirs(out_dir)
```

This **gene information file (gene_anno_file)** is sourced from the scFoundation project and contains gene token metadata.

```python
gene_list_df = pd.read_csv(gene_anno_file, header=0, delimiter='\t')
gene_list = list(gene_list_df['gene_name'])
```

To construct the single-cell data preprocessing object, it is recommended to use **scRNA_process_for_scFoundation.py** to preprocess the count file. This involves filtering low-quality cells and retaining only the genes utilized by scFoundation.<br>
If this step is skipped, please set **pre_normalized=False**.

```python
scp = scPreProcessing()
scp.set_data(sc_count_file=sc_count_file,sc_anno_file=sc_anno_file,pre_normalized=False)
scp.cell_selection()
scp.preprocessing_forFoundation(gene_names=gene_list)
```

**Optional**: Genes used here were selected via **scp.geneWords (with gene_use='HVG')**, consistent with the gene sets used in STED-CorEx and STED-LDA. See **tutorial.ipynb** for specific filtering rules.

```python
use_genes = pd.read_csv(os.path.join(data_dir,"use_gene.csv"),index_col=0).iloc[:,0].to_list()
scp.use_genes=use_genes
```

# Topic Extraction with scTopic #

Constructing the scTopicBert object for model training.

```python
from STED.scTopicBert import *
```

```python
sct = scTopicBERT()
sct.setData(scp,out_dir)
```

**Option 1**: Integrated approach The scTopicBert object can directly perform inference on single-cell data using a pre-trained model to generate embeddings.<br>
<br>
**Option 2**: Optimized workflow (Recommended) It is recommended to use **get_embedding.py** from the scFoundation project to pre-generate the embedding file. This approach allows for more extensive parameter tuning and allows you to skip the getEmbeddingModel step within this pipeline.<br>
<br>
Extension to Bulk Epigenomics: Similarly, if you provide a pre-computed gene activity score file for the target bulk epigenomic samples, the subsequent processing steps will be significantly simplified.<br>

```python
! python your_path/scFoundation/model/get_embedding.py --task_name PBMC_sc --input_type singlecell --output_type cell --pool_type all --tgthighres a5 --data_path PBMC_Main_demo.h5ad --save_path ./ --pre_normalized F --version rde
```

```python
! python your_path/scFoundation/model/get_embedding.py --task_name PBMC_gs --input_type bulk --output_type cell --pool_type all --tgthighres a5 --data_path PBMC_gs_Bulk_demo.h5ad --save_path ./ --pre_normalized F --version rde
```

```python
# This pretrained model is also from scFoundation project.
sct.getEmbeddingModel(model_type='scFoundation',pretrain_file = pretrained_model_file,tgthighres = "a5")
```

```python
# generating cell embeddings
sct.getEmbedding()
```

**c-TF-IDF (Class-based TF-IDF)**: Unlike standard TF-IDF, c-TF-IDF treats each cell cluster as a "document." it calculates gene importance relative to the cluster, helping to identify the most representative biological markers for each group.<br>
<br>
Feature Selection: By pre-defining use_genes, you filter out background noise (like housekeeping genes) so the model can focus on the signals that drive cellular identity.

```python
sct.getClusterModel(supervise = False,
                    cluster_method="cluster",
                    resolution=resolution,
                    min_samples=3,
                    min_cluster_size=int(len(sct.cells)*fc),
                    metric='euclidean', 
                    cluster_selection_method='eom',
                    n_neighbors=10,
                    n_components=2,
                    seed=seed)

sct.getCtfidfModel(bm25_weighting=True,
                   reduce_frequent_words=False)
```

```python
sct.fit(use_bin = False)
```

Alternatively, you can set the embedding_file, thus eliminating the need to run sct.getEmbedding()

```python
# sc_embedding_file = os.path.join(data_dir,"PBMC_Main_01B-resolution_singlecell_cell_embedding_a5_resolution.npy")
# sct.fit(use_bin = False,embedding_file=sc_embedding_file)
```

If a pre-computed gs_embedding_file is not available, a gsProcessing object must be constructed to preprocess the bulk samples (consistent with the STED-CorEx workflow).

Pre-normalized Setting: If the input consists of Gene Activity Score (GAS) files that have already been preprocessed via scRNA_process_for_scFoundation.py, please ensure that pre_normalized=True is set.

Additional Configurations: For all other parameter settings and detailed configurations, please refer to tutorial.ipynb.

```python
epid = gsPreProcessing()
epid.set_data(gs_count_file=gs_count_file,pre_normalized=False)
epid.preprocessing_forFoundation(gene_list)
```

```python
_,p_n=sct.transform(input_df=epid.target_df,similarity="cluster",device_name="cuda:0")
```

If the bulk sample embeddings have been pre-generated using get_embedding.py, you can directly execute the transform function to obtain the estimated cell proportions p_n. If you have both the gs_embedding_file and the sc_embedding_file, you don't even need to run sct.getEmbeddingModel().
